***ในแล็บนี้เราจะศึกษาวิธีการใช้ Language Model อย่าง N-Gram ในการสร้างประโยค***

Language Model เป็นโมเดลในการทำนายคำที่จะปรากฎคำต่อไป หรือบอกโอกาสที่จะเจอลำดับของคำที่กำหนด โดย Language Model ง่าย ๆ อันหนึ่งที่นิยมใช้กันคือ N-Gram โดยที่ `gram` บ่งบอกจำนวนคำที่มีการกำหนดค่าความน่าจะเป็น เช่น N-2 เรียกว่า Bigram คือเราจะกำหนดค่าความน่าจะเป็นให้กับวลีที่ประกอบไปด้วย 2 คำเป็นต้น

## Bigram

Bigram กำหนดค่าความน่าจะเป็นให้กับวลีที่มี 2 คำ ดังนั้นถ้าจะใช้ในการทำนายคำถัดไป ก็ต้องใช้คำก่อนหน้าหนึ่งคำในการทำนาย โดยในตัวอย่างนี้เราจะเริ่มจากคำว่า `แมว` แล้วใช้ Bigram ต่อคำไปเรื่อย ๆ จนกระทั่งได้ประโยค โดยเราจะใช้ `Bigram` ที่ถูกสร้างไว้แล้ว โหลดโดยการเรียก [`Bigram`](https://pythainlp.github.io/dev-docs/api/generate.html#pythainlp.generate.Bigram) แล้วเรียกฟังก์ชัน [`gen_sentence`](https://pythainlp.github.io/dev-docs/api/generate.html#pythainlp.generate.Bigram.gen_sentence) เพื่อต่อคำ

ลองเปลี่ยนคำดู แล้วดูว่าผลลัพธ์ที่ได้เป็นอย่างไร ถ้าต้องการประโยคที่ยาวขึ้นให้เพิ่มค่า `N`

```python
gen = Bigram()
```

```python
gen.gen_sentence("แมว", N=5)
```
